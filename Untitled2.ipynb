```python
#preprocessing phase
import pandas as pd
dataset = pd.read_csv("titanic-passengers.csv")
dataset.head()
```

```python
print(dataset["Cabin"].head())
dataset["Cabin"].head().isnull()
```

```python
print(dataset.head())
dataset["Cabin"].isnull().sum()
```

```python
dataset["Age"] = dataset["Age"].fillna(dataset["Age"].mean()).apply(lambda x: round(x, 2))
dataset.drop(['Cabin'], axis = 1, inplace = True)
dataset
```

```python
from sklearn.preprocessing import LabelEncoder

my_dict = {
    "male" : 0,
    "female" : 1,
}

my_array = []

def getSex(x):
    my_array.append(my_dict[x])

dataset['Sex'].apply(getSex)
sexSerie = pd.Series(my_array)
dataset["Sex"] = sexSerie

dataset
```

```python
dataset.drop(['Ticket'], axis = 1, inplace = True)
dataset
```

```python
import itertools
import operator

def most_common(L):
    l_to_list = sorted([str(x) for x in list(L.values)] ) 
    counts = [l_to_list.count(x) for x in l_to_list]
    index_max = counts.index(max(counts))
    return l_to_list[index_max]


# dataset
```

```python
dataset["Embarked"] = dataset["Embarked"].fillna(most_common(dataset["Embarked"]))
```

```python
from sklearn.preprocessing import LabelEncoder
my_dict = {
    "S" : 1,
    "C" : 2,
    "Q" : 3,
}

my_array = []

def getEmbark(x):
    my_array.append(my_dict[x])

dataset['Embarked'].apply(getEmbark)
embarkSerie = pd.Series(my_array)
dataset['Embarked'] = embarkSerie

dataset
```

```python
from sklearn.preprocessing import LabelEncoder

my_dict = {
    "No" : 0,
    "Yes" : 1,
}

my_array = []

def getSurvived(x):
    my_array.append(my_dict[x])

dataset['Survived'].apply(getSurvived)
survivedSerie = pd.Series(my_array)
dataset["Survived"] = survivedSerie

dataset
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.FacetGrid(dataset, col='Sex').map(plt.hist, "Survived")
dataset.head()
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(dataset)

# This graph makes a correlation beetween each variables. The corelation beetween variables which are close are near to one and inversely.
```

```python
#dataset= dataset.assign(Title=pd.Series(Title_Dictionary).values)
def get_title(name):
    Title_Dictionary = {

                        "Capt":       "Officer",

                        "Col":        "Officer",

                        "Major":      "Officer",

                          "Dr":         "Officer",

                        "Rev":        "Officer",

                        "Jonkheer":   "Royalty",

                        "Don":        "Royalty",

                        "Sir" :       "Royalty",

                       "Lady" :      "Royalty",

                      "the Countess": "Royalty",

                        "Dona":       "Royalty",

                        "Mme":        "Miss",

                        "Mlle":       "Miss",

                        "Miss" :      "Miss",

                        "Ms":         "Mrs",

                        "Mr" :        "Mrs",

                        "Mrs" :       "Mrs",

                        "Master" :    "Master"

                        }
    for k, v in Title_Dictionary.items():
        if k in name:
            return v
        
    return ''



dataset["Title"] = dataset["Name"].apply(get_title)
dataset
```
